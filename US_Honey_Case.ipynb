**Top Honey Producing States**

```python
df2 = df.sort_values("production",ascending=False)
```

```python
df2
```

```python
plt.figure(figsize=(20,10))

sns.barplot(x=df2["state"],y=df2["production"])
plt.xlabel("Total Honey Production")
plt.ylabel("Honey Producing states")
plt.xticks(rotation=90)
plt.title("States vs Honey Production")

plt.show()
```

```python
# from sys import float_repr_style
# from the above graph we can infer that the top honey producing states are:


# 1) North dakota
# 2) california
# 3) south dakota
# 4) florida
# 5) Montana
```

**Change in mean Average price of Honey From 1995 to 2021**

```python
df3 = df.groupby("year").mean()
```

```python
df3
```

```python
df3 = df3.reset_index()
```

```python
df3
```

```python
plt.figure(figsize=(20,10))

sns.barplot(x=df3["year"],y=df3["average_price"])
plt.xlabel("Year(1995-2021)")
plt.ylabel("Average Price")
plt.xticks(rotation=90)
plt.title("Year vs Mean Average Price")

plt.show()
```

**Which was the year when production of Honey was the highest?**

```python
plt.figure(figsize=(20,10))

sns.barplot(x=df["year"],y=df["production"])
plt.xlabel("Year(1995-2021)")
plt.ylabel("Production of Honey")
plt.xticks(rotation=90)
plt.title("Year vs Production of Honey")

plt.show()
```

```python
# from the above graph we can infer that the production of honey is maximum in year 2000 and minimum in 2021
```

**Which was the year when total mean value of production of Honey was the highest ?**

```python
df3
```

```python
plt.figure(figsize=(20,10))

sns.barplot(x=df3["year"],y=df3["value_of_production"])
plt.xlabel("Year(1995-2021)")
plt.ylabel("Value of Production of Honey")
plt.xticks(rotation=90)
plt.title("Year vs Mean Value of Production of Honey")

plt.show()
```

```python
plt.figure(figsize=(10,15))
plt.pie(df3["value_of_production"],labels=df3["year"],autopct="%.2f%%")
plt.show()
```

```python
# from the above graphs we can infer that the mean value of production of honey was maximum in year 2014 and minimum
# in year 1999
```
